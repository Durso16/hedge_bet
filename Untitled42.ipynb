```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Funo para calcular as apostas ideais para Time A e Time B, garantindo 3x o retorno e ajustando a aposta no empate
def calcular_apostas(odds_a, odds_b, odds_empate, max_investimento=50):
    # Calcula o investimento necessrio para garantir 3x o retorno em Time A e Time B
    investimento_a = max_investimento / (1 + (odds_b / odds_a) + (odds_empate / odds_a))
    investimento_b = (odds_b / odds_a) * investimento_a

    # Verifica quanto resta para apostar no empate
    investimento_empate = max_investimento - (investimento_a + investimento_b)

    # Se o valor para o empate for negativo, redistribumos as apostas
    if investimento_empate < 0:
        investimento_empate = 0
        # Recalcula as apostas em A e B para ficar dentro do oramento total
        investimento_a = max_investimento / (1 + (odds_b / odds_a))
        investimento_b = max_investimento - investimento_a

    return investimento_a, investimento_b, investimento_empate

# Funo para calcular o lucro/perda em cada cenrio
def calcular_lucro(aposta_a, aposta_b, aposta_empate, odds_a, odds_b, odds_empate):
    retorno_a = aposta_a * odds_a
    retorno_b = aposta_b * odds_b
    retorno_empate = aposta_empate * odds_empate

    # Lucro/perda considerando o valor total investido
    lucro_a = retorno_a - (aposta_a + aposta_b + aposta_empate)
    lucro_b = retorno_b - (aposta_a + aposta_b + aposta_empate)
    lucro_empate = retorno_empate - (aposta_a + aposta_b + aposta_empate)

    return lucro_a, lucro_b, lucro_empate

# Funo para simular diferentes cenrios e plotar os resultados
def simular_cenario(odds_a=2.16, odds_b=3.55, odds_empate=2.56):
    aposta_a, aposta_b, aposta_empate = calcular_apostas(odds_a, odds_b, odds_empate)
    lucro_a, lucro_b, lucro_empate = calcular_lucro(aposta_a, aposta_b, aposta_empate, odds_a, odds_b, odds_empate)

    # Definindo os resultados em ordem desejada: Time A, Empate, Time B
    resultados = ['Vitria Time A', 'Empate', 'Vitria Time B']
    lucros = [lucro_a, lucro_empate, lucro_b]

    # Grfico dos lucros/perdas
    plt.figure(figsize=(10, 6))
    plt.plot(resultados, lucros, marker='o', linestyle='-', color='b')
    plt.title('Simulao de Estratgia de Hedge')
    plt.xlabel('Cenrio')
    plt.ylabel('Lucro/Perda (R$)')
    plt.grid(True)
    plt.axhline(0, color='r', linestyle='--')
    plt.show()

    print(f'Aposta no Time A: R${aposta_a:.2f}')
    print(f'Aposta no Time B: R${aposta_b:.2f}')
    print(f'Aposta no Empate: R${aposta_empate:.2f}')
    print(f'Lucro se o Time A vencer: R${lucro_a:.2f}')
    print(f'Lucro se o Time B vencer: R${lucro_b:.2f}')
    print(f'Perda se houver empate: R${lucro_empate:.2f}')

# Interatividade com sliders para ajustar as odds
interact(simular_cenario,
         odds_a=FloatSlider(value=2.16, min=1.1, max=10.0, step=0.01, description='Odds Time A'),
         odds_b=FloatSlider(value=3.55, min=1.1, max=10.0, step=0.01, description='Odds Time B'),
         odds_empate=FloatSlider(value=2.56, min=1.1, max=10.0, step=0.01, description='Odds Empate'));



```
